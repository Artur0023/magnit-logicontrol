```python
import pandas as pd
import matplotlib.pyplot as plt

from analytics.src.db import load_table
```

```python
shipments = load_table("shipments")
dc = load_table("distribution_centers")
sku = load_table("sku")
suppliers = load_table("suppliers")

shipments.head()
```

```python
# % задержанных поставок

shipments["is_delayed"] = shipments["delay_days"] > 0

delay_rate = shipments["is_delayed"].mean() * 100
delay_rate
```

```python
# График

delay_by_dc = (
    shipments
    .groupby("dc_id")["is_delayed"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure()
delay_by_dc.plot(kind="bar")
plt.title("Доля задержанных поставок по РЦ")
plt.ylabel("Share")
plt.xlabel("DC")
plt.show()
```
